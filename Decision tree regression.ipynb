```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('petrol_consumption.csv')
```

```python
df
```

```python
df.corrwith(df['Petrol_Consumption'])
```

<h4>Separate the input output variable</h4>

```python
x=df[['Petrol_tax','Average_income','Paved_Highways','Population_Driver_licence(%)']]
y=df['Petrol_Consumption']
```

```python
sns.kdeplot(y);
```

```python
sns.kdeplot(x['Average_income'])
```

```python
sns.kdeplot(x['Paved_Highways'])
```

<h4>Cross-validation</h4>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train
```

```python
x_test
```

<h4>Build the model</h4>

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg=DecisionTreeRegressor(random_state=0)
```

```python
reg.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(30,14))
plot_tree(reg,fontsize=7,  filled=True, rounded=True, class_names=['O','F'],
         feature_names=['tax','inc','hways','popl']);
```

<h4>Performance Evaluation</h4>

```python
y_pred=reg.predict(x_test)
```

```python
result=pd.DataFrame({
    'actual':y_test,
    'predicted':y_pred
})
```

```python
result
```

```python
result['actual'].mean()
```

```python
mae = (result['actual'] - result['predicted']).abs().mean()
```

```python
mae
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
