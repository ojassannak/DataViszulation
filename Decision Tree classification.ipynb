Seprate  the input and output variables

```python
# input
x = df[["Variance","Skewness","Curtosis", "Entropy"]]
# Output
y = df["Class"]
```

<h4>Cross Validation</h4>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train
```

<h4>Build the Decision tree Model</h4>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(random_state=0)
```

```python
dt.fit(x_train,y_train)
```

<h4>Plot the tree</h4>

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(18,14))
plot_tree(dt,fontsize=6, filled=True, rounded=True, class_names=['O','F'],
         feature_names=x.columns);
```

<h4>Performance Evaluation</h4>

```python
y_pred=dt.predict(x_test)
```

```python
y_pred
```

```python
result=pd.DataFrame({
    'actual':y_test,
    'predicted':y_pred
})
```

```python
result
```

```python
(result['actual']==result['predicted']).sum()
```

```python
(result['actual']!=result['predicted']).sum()
```

```python
(result['actual']==result['predicted']).sum()/len(y_test)
```

```python
y_test.value_counts()
```

```python
sum(y_test[y_test==0]==y_pred[y_test==0])
```

```python
sum(y_test[y_test==1]==y_pred[y_test==1])
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```
