```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:/Users/duasp/datasets/mtcars.csv')
```

<h4>seprarate the input and output variable</h4>

```python
#input -> dis,hp,wt
x=df[['disp','hp','wt']]
#output-> mpg
y=df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
x.corrwith(y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
disp=221
hp=102
wt=2.91
```

```python
mileage=reg.predict([[disp,hp,wt]])
```
