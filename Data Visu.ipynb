```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([4,11,15,21,28,34,39,35,47,55])
y=np.array([10,22,31,36,48,51,58,61,69,77])
z=np.array([92,83,89,56,61,56,48,41,26,19])
p=np.array([56,44,27,65,38,61,23,75,36,58])
```

<h4>Line plot</h4>

```python
plt.plot(x,p)
```

```python
plt.title('My first plot')
plt.grid()
plt.xlim(0,60)
plt.ylim(0,100)
plt.xticks(range(0,61,5))
plt.yticks(range(0,101,5))
plt.xlabel('First Value')
plt.ylabel('Second Value')
# markers: o  .  <  >  v  ^  *   D   +
#ls: : - -.
plt.plot(x,y,color='#000000',marker='D',lw=1,mfc='red',ms=5,ls=':')
```

```python
plt.title('My first comapre plot')
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x,y,color='#808000',label='2022')
plt.plot(x,z,color='#000000',label='2023')
plt.legend()
```

<h4>Bar Plot</h4>

```python
enrol=pd.DataFrame({
    'Students':[67,89,106,179,47,122,100],
    'Languages':['C','C++','Java','Python','R','PHP','Javascript']
})
```

```python
enrol
```

```python
plt.title('Students Enrollment for Training')
plt.bar(enrol['Languages'],enrol['Students'],color='green')
plt.xlabel('Languages')
plt.ylabel('Total Students')
```

```python
plt.title('Students Enrollment for Training')
plt.bar(enrol['Languages'],enrol['Students'],color='blue')
plt.xlabel('Languages')
plt.ylabel('Total Students')
plt.plot(enrol['Languages'],enrol['Students'],marker='D')
```

```python
plt.title('Students Enrollment for Training')
plt.barh(enrol['Languages'],enrol['Students'],color='brown')
plt.ylabel('Languages')
plt.xlabel('Total Students')
```

```python
enrol=pd.DataFrame({
    'Students':[67,89,106,179,47,122,100],
    'Languages':['C','C++','Java','Python','R','PHP','Javascript'],
    'Avgmarks':[67.45,69.23,72.45,78.23,63.27,69.34,74.23]
})
```

```python
enrol
```

```python
plt.title('Students Enrollment for Training')
plt.bar(enrol['Languages'],enrol['Students'],color='Navy',align='edge',width=0.3,label='Total Students')
plt.bar(enrol['Languages'],enrol['Avgmarks'],color='Gray',align='center',width=0.2,label='Avgmarks')


plt.xlabel('Languages')
plt.ylabel('Total Students')
plt.legend()
```

<h4>Pie Chart</h4>
